## Seeding the Generator
(https://github.com/joke2k/faker#seeding-the-generator)

Ïª¥Ìì®ÌÑ∞ ÌîÑÎ°úÍ∑∏ÎûòÎ∞çÏóêÏÑú ÏûÑÏùòÏùò Í∞íÏùÑ Î∞òÌôòÌïòÎäî Í≤ΩÏö∞(ÎÇúÏàò ÏÉùÏÑ± Îì±) ÏãúÎìúÎùºÎäî Í∞úÎÖêÏù¥ ÏûàÎã§.<br>
ÏãúÎìúÎ•º ÏÑ§Ï†ïÌïòÍ≤å ÎêòÎ©¥ ÎèôÏùºÌïú ÏàúÏÑúÎ°ú ÎÇúÏàòÎ•º Î∞úÏÉùÏãúÌÇ¨ Ïàò ÏûàÏñ¥ ÏùºÎ∞òÏ†ÅÏúºÎ°ú ÎîîÎ≤ÑÍπÖÏùÑ
ÏúÑÌïòÏó¨ ÌôúÏö© ÎêúÎã§.


````python
import random

random.random()    # => ÏûÑÏùòÏùò Ïàò
random.random()    # => ÏûÑÏùòÏùò Ïàò

random.seed(7777)
random.random()    # => 0.8170477907294282

random.seed(7777)
random.random()    # => 0.8170477907294282
````

ÏïÑÎûòÏùò ÏΩîÎìúÎ•º Ïã§Ìñâ ÌñàÏùÑ Îïå, #1Í≥º #2ÏóêÏÑú Ï∂úÎ†•ÎêòÎäî Í≤∞Í≥ºÎ•º Í∞ÅÍ∞Å ÏûëÏÑ±ÌïòÍ≥†,<br> seed()Îäî
Ïñ¥Îñ§ Ï¢ÖÎ•òÏùò Î©îÏÑúÎìúÏù∏ÏßÄ ÏûëÏÑ±ÌïòÏãúÏò§.

````python
from faker import Faker   


fake = Faker('ko_KR')
Faker.seed(4321)

print(fake.name())    # 1

fake2 = Faker('ko_KR')
print(fake2.name())   # 2
````

**seed()** <br>
: ÎÇúÏàòÎ•º ÏÉùÏÑ±Ìï† Îïå ÏùºÏ¢ÖÏùò Í∏∞Ï§ÄÏù¥ Ï°¥Ïû¨ÌïúÎã§.<br>
Ïù¥Í≤ÉÏùÑ Ï†ïÌï¥Ï£ºÎäî Ïó≠Ìï†Ïù¥ seed() Ìï®ÏàòÏù¥Îã§!<br><br>

ÏãúÎìúÍ∞Ä Í∞ôÏùÑ Í≤ΩÏö∞ÏóêÎäî Îã§Ïùå ÏÉùÏÑ±Îê† ÎÇúÏàòÍ∞Ä ÏòàÏ∏°ÎêúÎã§Îäî Í≤ÉÏù¥ Ï§ëÏöîÌïòÎã§!

## Seeding the Generator

ÏïÑÎûòÏùò ÏΩîÎìúÎ•º Ïã§Ìñâ ÌñàÏùÑ Îïå, #1Í≥º #2ÏóêÏÑú Ï∂úÎ†•ÎêòÎäî Í≤∞Í≥ºÎ•º Í∞ÅÍ∞Å ÏûëÏÑ±ÌïòÍ≥†,
seed_instance()Îäî Ïñ¥Îñ§ Ï¢ÖÎ•òÏùò Î©îÏÑúÎìúÏù∏ÏßÄ ÏûëÏÑ±ÌïòÏãúÏò§.

````python
from faker import Faker 

fake = Faker('ko_KR')
fake.seed_instance(4321)

print(fake.name())    # 1

fake2 = Faker('ko_KR')
print(fake2.name())   # 2
````

`seed()`ÏôÄ `seed_instance()`Îäî Í∞ÅÍ∞Å Ïñ¥Îñ†Ìïú Ïö©ÎèÑÎ°ú Ïì∞Ïùº Ïàò ÏûàÎäîÏßÄ ÏûëÏÑ±ÌïòÏãúÏò§.

**seed()**<br>
