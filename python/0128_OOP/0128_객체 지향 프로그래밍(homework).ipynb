### Circle ì¸ìŠ¤í„´ìŠ¤ ë§Œë“¤ê¸°

ì•„ë˜ì™€ ê°™ì€ Circle í´ë˜ìŠ¤ê°€ ìˆì„ ë•Œ, <br>
ë°˜ì§€ë¦„ì´ 3ì´ê³  x, yì¢Œí‘œê°€ (2, 4)ì¸ Circle ì¸ìŠ¤í„´ìŠ¤ë¥¼ ë§Œë“¤ì–´<br>
ë„“ì´ì™€ ë‘˜ë ˆë¥¼ ì¶œë ¥í•˜ì‹œì˜¤.

```python
class Circle:
    pi = 3.14
    x = 0
    y = 0
    r = 0
    
    def __init__(self, r, x, y):
        self.r = r
        self.x = x
        self.y = y
        
    def area(self):
        return self.pi * self.r * self.r
    
    def circumference(self):
        return 2 * self.pi * self.r
    
    def center(self):
        return (self.x, self.y)
```

```python
# ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
c1 = Circle(3, 2, 4)

# ë„“ì´
print("ë„“ì´:", c1.area())

# ë‘˜ë ˆ
print("ë‘˜ë ˆ:", c1.circumference())
```
