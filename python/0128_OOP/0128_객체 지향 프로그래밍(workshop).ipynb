## 도형 만들기

### 나의 답

#### Point  클래스 생성

````python
class Point:
    
    # 생성자
    def __init__(self, x, y):
        # 인스턴스 변수
        self.x = x
        self.y = y
````

#### Rectangle 클래스 생성

````python
class Rectangle():
    
    # 생성자
    def __init__(self, p1, p2):
        # 인스턴스 변수 - Point 인스턴스 
        # p1과 p2는 단순한 int가 아니라, 하나의 Point 인스턴스가 들어간다.
        self.p1 = p1
        self.p2 = p2
    
    # 넓이
    def get_area(self):
        # self.p1.x = r1.p1.x : r1 = Rectangle()로 인스턴스 생성 시,
        area = abs(self.p1.x - self.p2.x) * abs(self.p1.y - self.p2.y)
        return area
    
    # 둘레 길이
    def get_perimeter(self):
        perimeter = 2 * (abs(self.p1.x - self.p2.x) + abs(self.p1.y - self.p2.y))
        return perimeter
    
    # 정사각형 유무
    def is_square(self):
        real_square = abs(self.p1.x - self.p2.x) == abs(self.p1.y - self.p2.y)
        return real_square
````

#### 테스트

````python
p1 = Point(1, 3)
p2 = Point(3, 1)
r1 = Rectangle(p1, p2)

print(r1.get_area())
print(r1.get_perimeter())
print(r1.is_square())

p3 = Point(3, 7)
p4 = Point(6, 4)
r2 = Rectangle(p3, p4)

print(r2.get_area())
print(r2.get_perimeter())
print(r2.is_square())
````
