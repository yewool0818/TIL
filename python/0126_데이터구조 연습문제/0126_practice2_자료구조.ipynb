````python
def my_find(text, alphabet):
    
    # ì œì‹œëœ ì•ŒíŒŒë²³ì´ ì£¼ì–´ì§„ ë¬¸ìì—´ì— ë“±ì¥í•˜ì§€ ì•Šìœ¼ë©´,
    if alphabet not in text:
        return -1 # -1 ë°˜í™˜ ë !
    
    
    # ì œì‹œëœ ì•ŒíŒŒë²³ì´ ì£¼ì–´ì§„ ë¬¸ìì—´ì— ë“±ì¥í•˜ë©´,
    result = []
     
    for idx, char in enumerate(text):
        if char == alphabet:
            result.append(idx)
    
    return result
    
````

````python
def check(n, students):
    
    #ê²°ì„ í•™ìƒì„ ë°˜í™˜í•´ì¤„ ë¬¸ìì—´ ì´ˆê¸°í™”
    absent_students = ''
    
    # ì´ ëª¨ë“  í•™ìƒ ëª…ë¶€ ë¦¬ìŠ¤íŠ¸
    # - ë¦¬ìŠ¤íŠ¸ê°€ ì•„ë‹Œ ë¬¸ìì—´ë¡œ ë§Œë“¤ì–´ ì¤€ë‹¤.
    all_students = map(str, list(range(1, n+1))) #['1','2','3','4','5', '6', '7']
    
    # all_studentsë¥¼ ìˆœíšŒí•˜ë©°, 
    for all_student in all_students:
        # students(ì¶œì„í•œ í•™ìƒ)ì— ì—†ìœ¼ë©´, ê²°ì„ í•™ìƒìœ¼ë¡œ ì¶”ê°€
        # studentsëŠ” ë¬¸ìì—´ì´ê¸° ë•Œë¬¸ì—, ê³µë°±ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ í˜•ì‹ìœ¼ë¡œ ë§Œë“¤ì–´ì¤€ë‹¤.
        if all_student not in students.split():
            absent_students += all_student + ' '
        
    return absent_students
````

````python
# í•´ë‹¹ ì½”ë“œë¥¼ í†µí•´ ì˜¬ë°”ë¥¸ ê²°ê³¼ê°€ ë‚˜ì˜¤ëŠ”ì§€ í™•ì¸í•˜ì‹œì˜¤.
print(check(7, '1 3 5')) # 2 4 6 7
````

### ğŸ‘¨â€ğŸ« Solution
argumentê°€ Stringìœ¼ë¡œ ë“¤ì–´ì™”ì„ ë•Œ, í‘¸ëŠ” ë°©ë²•ì„ ì—°ìŠµí•´ë³´ì!

````python
def check(n, students):
    
    # ì¶œì„í•œ í•™ìƒ : '1 3 5' -> [1, 3, 5]
    students = list(map(int, students.split())) #.split()ì€ ê¸°ë³¸ìœ¼ë¡œ strìœ¼ë¡œ ë°˜í™˜ë˜ì–´ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥ëœë‹¤.
    
    # ê²°ì„ì ì¶œì„ ë²ˆí˜¸ ë¦¬ìŠ¤íŠ¸
    absents = []
    
    # 1ë²ˆë¶€í„° në²ˆê¹Œì§€ (ì „ì²´ í•™ìƒ) ì¶œì„ ë²ˆí˜¸ ìˆœíšŒ
    for i in range(1, n+1):
        # ië²ˆ í•™ìƒì´ ì¶œì„í•œ í•™ìƒ ë¦¬ìŠ¤íŠ¸ì— ì—†ìœ¼ë©´,
        if i not in students:
            absents.append(str(i))# ê²°ì„ì ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    
    # '[]'.join(list) : list -> string
    result = ' '.join(absents)
    
    return result
    
````
